```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
total_schools = school_data_complete["school_name"].nunique()
total_students = school_data_complete["Student ID"].count()
school_data_complete_new = school_data_complete.groupby(["school_name"]).max()
total_budget = school_data_complete_new["budget"].sum()
avg_math_score = school_data_complete["math_score"].mean()
avg_reading_score = school_data_complete["reading_score"].mean()

tot_pass_math = school_data_complete.loc[school_data_complete["math_score"] >= 70]
percent_pass_math = (tot_pass_math["math_score"].count()/total_students)*100

tot_pass_reading = school_data_complete.loc[school_data_complete["reading_score"] >= 70]
percent_pass_reading = (tot_pass_reading["reading_score"].count()/total_students)*100

percent_pass_overall = (avg_math_score + avg_reading_score)/2

df_district_summary = pd.DataFrame({"Total Schools":[total_schools],"Total Students":[total_students],"Total Budget":[total_budget],"Average Math Score":[avg_math_score],
                                   "Average Reading Score":[avg_reading_score],"% Passing Math":[percent_pass_math],"% Passing Reading":[percent_pass_reading],"% Overall Passing Rate":[percent_pass_overall]})
df_district_summary.style.format({"Total Schools":"{:,}","Total Students":"{:,}","Total Budget":"${:.2f}","Average Math Score"
                                 :"{:.2f}","Average Reading Score":"{:.2f}","% Passing Math":"{:.2f}","% Passing Reading"
                                 :"{:.2f}","% Overall Passing Rate":"{:.2f}"})
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
df = school_data_complete.groupby(["school_name"])
school_type = df["type"].max()
total_student_count = df["student_name"].count()
total_school_budget = df["budget"].max()

per_student_budget = total_school_budget/total_student_count
avg_math_score = df["math_score"].mean()
avg_reading_score = df["reading_score"].mean()


tot_pass_math = school_data_complete.loc[school_data_complete["math_score"] >= 70]
tot_pass_math_per_school = tot_pass_math.groupby(["school_name"])["math_score"].count()
percent_pass_math = (tot_pass_math_per_school/total_student_count)*100

tot_pass_reading = school_data_complete.loc[school_data_complete["reading_score"] >= 70]
tot_pass_reading_per_school = tot_pass_reading.groupby(["school_name"])["reading_score"].count()
percent_pass_reading = (tot_pass_reading_per_school/total_student_count)*100


overall_pass_rate = (percent_pass_math + percent_pass_reading)/2

df_school_summary = pd.DataFrame({"School Type":school_type,"Total Students":total_student_count,"Total School Budget":total_school_budget
                                  ,"Per Student Budget":per_student_budget,"Average Math Score":avg_math_score
                                  ,"Average Reading Score":avg_reading_score,"% Passing Math":percent_pass_math,"% Passing Reading":percent_pass_reading
                                  ,"Overall Passing Rate":overall_pass_rate})
df_school_summary
df_school_summary_sort = df_school_summary.sort_values("Overall Passing Rate",ascending = False)
df = df_school_summary_sort.head(5)
df.style.format({"School Type":"{:}","Total Students":"{:,}","Total School Budget":"${:.2f}","Per Student Budget":
                 "${:.2f}","Average Math Score":"{:.2f}","Average Reading Score":"{:.2f}","% Passing Math":"{:.2f}",
                 "% Passing Reading":"{:.2f}","Overall Passing Rate":"{:.2f}"})
```
